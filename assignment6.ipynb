```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Revenue": [45000,52000,48000,61000,58000,67000,72000,69000,75000,81000,88000,95000],
    "Expenses": [32000,35000,30000,40000,38000,42000,45000,44000,46000,50000,54000,58000],
    "Profit": [13000,17000,18000,21000,20000,25000,27000,25000,29000,31000,34000,37000]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"],
         df["Revenue"],
         linewidth=2,
         marker='o',
         color='steelblue')

plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Month"], df["Revenue"], color='blue', label='Revenue')

plt.plot(df["Month"], df["Expenses"], color='red', label='Expenses')

plt.plot(df["Month"], df["Profit"], color='green', label='Profit')

plt.title("Revenue vs Expenses vs Profit")

plt.xlabel("Month")
plt.ylabel("Amount")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Revenue"],
         color='steelblue',
         linewidth=2)

plt.fill_between(df["Month"],
                 df["Revenue"],
                 alpha=0.25,
                 color='steelblue')

plt.title("Revenue Area Chart")

plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.fill_between(df["Month"],
                 0,
                 df["Revenue"],
                 color='skyblue',
                 alpha=0.7,
                 label='Revenue')

plt.fill_between(df["Month"],
                 0,
                 df["Expenses"],
                 color='orange',
                 alpha=0.7,
                 label='Expenses')

plt.title("Revenue and Expenses Area Chart")
plt.xlabel("Month")
plt.ylabel("Amount")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
```

```python
np.random.seed(42)

ClassA = np.random.normal(70,10,50)
ClassB = np.random.normal(80,8,50)
ClassC = np.random.normal(65,12,50)

plt.figure(figsize=(8,6))

plt.boxplot([ClassA, ClassB, ClassC],
            labels=["Class A","Class B","Class C"],
            patch_artist=True)

plt.title("Box Plot of Three Classes")

plt.ylabel("Scores")

plt.grid(alpha=0.3)

plt.show()
```

## Observations

- Class B has the highest median score.
- Class C has the largest spread of scores.
- Any individual points outside the whiskers are outliers.
- The class with more visible outlier points has the most outliers.

```python
plt.figure(figsize=(12,6))

plt.plot(df["Month"],
         df["Revenue"],
         color='blue',
         linewidth=2,
         marker='o',
         label="Revenue")

plt.plot(df["Month"],
         df["Expenses"],
         color='red',
         linewidth=2,
         marker='o',
         label="Expenses")

highest = df["Revenue"].max()
month = df.loc[df["Revenue"].idxmax(), "Month"]

plt.annotate("Highest Revenue",
             xy=(month, highest),
             xytext=("Sep", 90000),
             arrowprops=dict(facecolor='black',
                             arrowstyle='->'))

plt.axvspan(6,8,
            color='yellow',
            alpha=0.3)

plt.title("Revenue and Expenses with Peak Season")

plt.xlabel("Month")

plt.ylabel("Amount")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
```
